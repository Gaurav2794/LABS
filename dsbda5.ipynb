```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.get_dummies(df, drop_first=True)
```

```python
x = data.drop(['Purchased'], axis = 1)
y = data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,random_state = 0)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
model = logreg.fit(xtrain,ytrain)
```

```python
ytrain_pred = logreg.predict(xtrain)
ytest_pred = logreg.predict(xtest)
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)
```

Execution with Formulas

Accuracy

```python
accuracy_train = np.sum(ytrain == ytrain_pred) / len(ytrain)
print(accuracy_train)
```

```python
accuracy_test = np.sum(ytest == ytest_pred) / len(ytest)
print(accuracy_test)
```

Error Rate

```python
error_train = 1 - (np.sum(ytrain == ytrain_pred) / len(ytrain))
print(error_train)
```

```python
error_test = 1 - (np.sum(ytest == ytest_pred) / len(ytest))
print(error_test)
```

Precision

```python
tp = np.sum((ytrain == 1) & (ytrain_pred == 1))
fp = np.sum((ytrain == 0) & (ytrain_pred == 1))
precision_train = tp / (tp + fp)
print(precision_train)
```

```python
tp = np.sum((ytest == 1) & (ytest_pred == 1))
fp = np.sum((ytest == 0) & (ytest_pred == 1))
precision_test = tp / (tp + fp)
print(precision_test)
```

Recall

```python
tp = np.sum((ytrain == 1) & (ytrain_pred == 1))
fn = np.sum((ytrain == 1) & (ytrain_pred == 0))
recall_train = tp / (tp + fn)
print(recall_train)
```

```python
tp = np.sum((ytest == 1) & (ytest_pred == 1))
fn = np.sum((ytest == 1) & (ytest_pred == 0))
recall_test = tp / (tp + fn)
print(recall_test)
```

F1 Score

```python
tp = np.sum((ytrain == 1) & (ytrain_pred == 1))
fp = np.sum((ytrain == 0) & (ytrain_pred == 1))
fn = np.sum((ytrain == 1) & (ytrain_pred == 0))
f1_train = (2 * tp) / (2 * tp + fp + fn)
print(f1_train)
```

```python
tp = np.sum((ytest == 1) & (ytest_pred == 1))
fp = np.sum((ytest == 0) & (ytest_pred == 1))
fn = np.sum((ytest == 1) & (ytest_pred == 0))
f1_test = (2 * tp) / (2 * tp + fp + fn)
print(f1_test)
```
