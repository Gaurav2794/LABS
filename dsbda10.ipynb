# Data Preprocessing

Importing libraries

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/iris.csv")
```

Adding outliers

```python
outliers = pd.DataFrame([
    [8.5,4.5,7.5,3.0,"virginica"],
    [2.5,1.5,0.5,0.1,"setosa"],
    [9.0,3.0,6.0,2.5,"versicolor"],
    [3.5,5.0,1.2,0.3,"setosa"],
    [7.8,2.0,5.5,1.0,"virginica"],
    [4.0,4.2,1.3,0.4,"setosa"],
    [6.0,1.5,4.0,0.8,"versicolor"],
    [5.5,3.8,6.5,2.2,"virginica"],
    [5.0,3.5,10.0,0.2,"setosa"],
    [5.0,0.5,1.4,0.2,"setosa"],
    [6.0,3.0,4.5,5.0,"versicolor"],
    [10.0,3.0,4.5,1.5,"versicolor"]
], columns=data.columns)

data = pd.concat([data, outliers], ignore_index=True)
```

Create a histogram for each feature in the dataset to illustrate the feature distributions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data["sepal_length"])
plt.title("sepal_length")
plt.show()
```

```python
sns.histplot(data["sepal_width"])
plt.title("sepal_width")
plt.show()
```

```python
sns.histplot(data["petal_length"])
plt.title("petal_length")
plt.show()
```

```python
sns.histplot(data["petal_width"])
plt.title("petal_width")
plt.show()
```

Create a boxplot for each feature in the dataset.

```python
sns.boxplot(x=data["sepal_length"])
plt.title("sepal_length")
plt.show()
```

```python
sns.boxplot(x=data["sepal_width"])
plt.title("sepal_width")
plt.show()
```

```python
sns.boxplot(x=data["petal_length"])
plt.title("petal_length")
plt.show()
```

```python
sns.boxplot(x=data["petal_width"])
plt.title("petal_width")
plt.show()
```

```python
sns.heatmap(data.select_dtypes(include=np.number).corr(), cmap = "YlGnBu", linecolor = 'white', linewidths = 1)
```
