# DSBDA Lab 12: Hadoop Log File Processing

This lab demonstrates how to process log files using Hadoop MapReduce.

### Start Hadoop Services
```bash
start-dfs.sh
start-yarn.sh
```

### Prepare HDFS
```bash
hdfs dfs -mkdir /loginput
hdfs dfs -put logfile.txt /loginput
```

### Compile Java Code
```bash
javac -classpath `hadoop classpath` -d log_classes LogProcessor.java
jar -cvf log.jar -C log_classes/ .
```

### Run MapReduce Job
```bash
hadoop jar log.jar LogProcessor /loginput /logoutput
```

### View Results
```bash
hdfs dfs -cat /logoutput/part-r-00000
```
