Step 1: Import libraries and create alias for Pandas, Numpy and Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Step 2: Import the Boston Housing dataset

```python
data = pd.read_csv("/content/BostonHousingData.csv")
```

```python
data
```

```python
data.head()
```

```python
data.fillna(0, inplace=True)
```

Split dependent variable and independent variables

```python
x = data.drop(['MEDV'], axis = 1)
y = data['MEDV']
```

Splitting data to training and testing dataset.

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,random_state = 0)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

Use linear regression( Train the Machine ) to Create Model

```python
import sklearn
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model=lm.fit(xtrain, ytrain)
```

Predict the y_pred for all values of train_x and test_x [Yp]

```python
ytrain_pred = lm.predict(xtrain)
ytest_pred = lm.predict(xtest)
```

Evaluate the performance of Model for train_y and test_y

```python
df=pd.DataFrame(ytrain_pred,ytrain)
df
```

```python
df=pd.DataFrame(ytest_pred,ytest)
df
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

Calculate Mean Absolute Error for train_y and test_y

```python
mae_test = mean_absolute_error(ytest, ytest_pred)
mae_train = mean_absolute_error(ytrain, ytrain_pred)
```

```python
mae_test
```

```python
mae_train
```

Calculate Mean Squared Error for train_y and test_y



```python
mse_test = mean_squared_error(ytest, ytest_pred)
mse_train = mean_squared_error(ytrain, ytrain_pred)
```

```python
mse_test
```

```python
mse_train
```

Calculate Root Mean Squared Error for train_y and test_y

```python
rmse_test = np.sqrt(mse_test)
rmse_train = np.sqrt(mse_train)
```

```python
rmse_test
```

```python
rmse_train
```

Calculate R2 Score for train_y and test_y

```python
r2_test = r2_score(ytest, ytest_pred)
r2_train = r2_score(ytrain, ytrain_pred)
```

```python
r2_test
```

```python
r2_train
```

Plotting the linear regression model

```python
plt.scatter(ytrain ,ytrain_pred,c='blue',marker='o',label='Training data')
plt.scatter(ytest,ytest_pred ,c='lightgreen',marker='s',label='Test data')
plt.xlabel('True values')
plt.ylabel('Predicted')
plt.title("True value vs Predicted value")
plt.legend(loc= 'upper left')
plt.plot()
plt.show()
```

```python
r2_test = 1 - (np.sum((ytest - ytest_pred)**2) / np.sum((ytest - np.mean(ytest))**2))
print(r2_test)
```
