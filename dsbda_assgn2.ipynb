# B. Identification and Handling of Outliers

Identification of Outliers
<br>
*an Outlier is an observation in a given dataset that lies far from the rest
of the observations. That means an outlier is vastly larger or smaller than the remaining values in the set.*

**Detecting Outliers**

**1. Detecting outliers using Boxplot:**

```python
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/student_marks_dataset.csv")
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (9, 6)
df_list = ['Rollno','Subject1_marks','Total_Marks','Percentage']
fig, axes = plt.subplots(2, 2)
fig.set_dpi(120)
count=0
for r in range(2):
  for c in range(2):
    _ = df[df_list[count]].plot(kind = 'box', ax=axes[r,c])
    count+=1
```

**Detecting outliers using Inter Quantile Range(IQR):**

```python
Q1 = df['Percentage'].quantile(0.25)
Q3 = df['Percentage'].quantile(0.75)
IQR = Q3 - Q1

Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR

print(f'Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}, Lower_limit = {Lower_limit}, Upper_limit = {Upper_limit}')
```

```python
df[(df['Percentage'] < Lower_limit) | (df['Percentage'] > Upper_limit)]
```

Z score

```python
from scipy import stats

z_scores = stats.zscore(df['Percentage'])

df['Z_score'] = z_scores

df[['Percentage', 'Z_score']]
```

```python
df[abs(df['Z_score']) > 3]
```

```python
df[abs(df['Z_score']) > 2]
```

removing the outlier:

```python
outliers=[]
for i in df.Percentage:
    if i<Lower_limit or i>Upper_limit:
        outliers.append(i)
print("outliers are",outliers)
```
