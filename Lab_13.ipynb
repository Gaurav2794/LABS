# DSBDA Lab 13: Scala Spark Sorting

Implement Bubble sort Algorithm using Scala programming.

### Scala Source Code
```scala
object BubbleSort {
  def bubbleSort(arr: Array[Int]): Array[Int] = {
    val n = arr.length
    for (i <- 0 until n - 1) {
      for (j <- 0 until n - i - 1) {
        if (arr(j) > arr(j + 1)) {
          // swap
          val temp = arr(j)
          arr(j) = arr(j + 1)
          arr(j + 1) = temp
        }
      }
    }
    arr
  }

  def main(args: Array[String]): Unit = {
    val arr = Array(64, 34, 25, 12, 22, 11, 90)
    println("Original Array: " + arr.mkString(", "))
    val sorted = bubbleSort(arr)
    println("Sorted Array: " + sorted.mkString(", "))
  }
}
```
