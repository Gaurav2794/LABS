# DSBDA Lab 11: Hadoop MapReduce WordCount

This lab demonstrates how to compile and run a Hadoop MapReduce WordCount program.

### Start Hadoop Services
```bash
start-dfs.sh
start-yarn.sh
```

### Prepare HDFS
```bash
hdfs dfs -mkdir /input
hdfs dfs -put input.txt /input
```

### Compile Java Code
```bash
javac -classpath `hadoop classpath` -d wc_classes WordCount.java
jar -cvf wc.jar -C wc_classes/ .
```

### Run MapReduce Job
```bash
hadoop jar wc.jar WordCount /input /output
```

### View Results
```bash
hdfs dfs -cat /output/part-r-00000
```
