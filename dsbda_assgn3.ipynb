Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable. For example, if your categorical variable is age groups and quantitative variable is income, then provide summary statistics of income grouped bythe age groups. Create a list that contains a numeric value for each response to the categorical variable.

```python
import pandas as pd
```

```python
dictionary = {
'AgeGroup': ['20-30', '20-30', '30-40', '30-40', '40-50', '40-50'],
'Income': [35000, 37000, 45000, 46000, 55000, 57000]
}
df = pd.DataFrame(dictionary)
```

```python
print(df)
```

```python
grouped = df.groupby('AgeGroup')['Income'].agg(['mean', 'median', 'min', 'max', 'var', 'std'])
grouped
```

```python
grouped = df.groupby('AgeGroup')['Income'].describe()
```

```python
print(grouped)
```

```python
grouped_list = df.groupby('AgeGroup')['Income'].apply(list).to_dict()
```

```python
print(grouped_list)
```

Part B: <br>
Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris- versicolor’ of iris.csv dataset.

```python
df = pd.read_csv("/content/iris.csv")
```

```python
print(df)
```

Statistics for Individual Species of Iris

```python
setosa = df['species'] == "setosa"
versicolor = df['species'] == "versicolor"
virginica = df['species'] == "virginica"
```

```python
print("Statisics for Setosa")
print(df[setosa].describe())
print("\nStatisics for Versicolor")
print(df[versicolor].describe())
print("\nStatisics for Virginica")
print(df[virginica].describe())
```

Statistics using For Loop

```python
for s in df['species'].unique():
  print(f"Statistics for {s}")
  print(df[df['species'] == s].describe())
```
