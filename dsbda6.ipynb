Import libraries and create alias for Pandas, Numpy and Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import the Iris dataset by calling URL.

```python
df = pd.read_csv("/content/iris.csv")
```

```python
df
```

```python
df.isnull().sum()
```

Converting Categorical to Numerical Values (Label Encoding)

```python
from sklearn import preprocessing
```

```python
df['species'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['species']= label_encoder.fit_transform(df['species'])
```

```python
df
```

```python
x = df.drop(['species'], axis = 1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,random_state = 0)
```

#Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
gaussian.fit(xtrain, ytrain)
```

```python
ytrain_pred = gaussian.predict(xtrain)
ytest_pred = gaussian.predict(xtest)
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)
```

Accuracy Score

```python
print(accuracy_score(ytrain, ytrain_pred))
```

```python
print(accuracy_score(ytest, ytest_pred))
```

Error Rate

```python
print(1-accuracy_score(ytrain, ytrain_pred))
```

```python
print(1-accuracy_score(ytest, ytest_pred))
```

Precision Score

```python
precision_score(ytrain, ytrain_pred,average='micro')
```

```python
precision_score(ytest, ytest_pred,average='micro')
```

Recall Score

```python
recall_score(ytrain, ytrain_pred,average='micro')
```

```python
recall_score(ytest, ytest_pred,average='micro')
```

Classification Report

```python
print(classification_report(ytrain, ytrain_pred))
```

```python
print(classification_report(ytest, ytest_pred))
```

Confusion Matrix

```python
print(confusion_matrix(ytrain, ytrain_pred))
```

```python
print(confusion_matrix(ytest, ytest_pred))
```

F1 Score

```python
f1_score(ytrain, ytrain_pred, average="micro")
```

```python
f1_score(ytest, ytest_pred,average='micro')
```

Precision

```python
tp = np.sum(ytrain == ytrain_pred)
fp = len(ytrain) - tp
precision_train = tp / (tp + fp)
print(precision_train)
```

```python
tp = np.sum(ytest == ytest_pred)
fp = len(ytest) - tp
precision_test = tp / (tp + fp)
print(precision_test)
```

Recall

```python
tp = np.sum(ytrain == ytrain_pred)
fn = len(ytrain) - tp
recall_train = tp / (tp + fn)
print(recall_train)
```

```python
tp = np.sum(ytest == ytest_pred)
fn = len(ytest) - tp
recall_test = tp / (tp + fn)
print(recall_test)
```

F1 Score

```python
tp = np.sum(ytrain == ytrain_pred)
fp = len(ytrain) - tp
fn = len(ytrain) - tp
f1_train = (2 * tp) / (2 * tp + fp + fn)
print(f1_train)
```
